```python
from drone_range_assignment import Solver, Instance, Point
```

```python
#instance = Instance([Point(0,0), Point(5, 3), Point(10, 4), Point(20, 20)], 1)
solutions = []
for n in range(4):
    instance = Instance([Point(0,0), Point(5, 10), Point(10,0)], n)
    model = Solver(instance)
    solution = model.solve()
    solutions.append(solution)
    assert solution.is_feasible(0.01)
```

```python
import matplotlib.pyplot as plt
from drone_range_assignment import plot
fix, axes = plt.subplots(1,len(solutions), figsize = (15,4))
for i in range(len(solutions)):
    ax = axes[i]
    solution = solutions[i]
    plot(ax, solution)
    ax.set_title(f"{i} drone(s) \nPower: {round(solution.objective(),2)}")
plt.tight_layout()
plt.savefig("../docs/figures/example.png")
plt.show()
```
