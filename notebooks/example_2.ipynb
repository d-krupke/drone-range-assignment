```python
from drone_range_assignment import Solver, Instance, Point
```

```python
instance = Instance([Point(0,0), Point(5, 3), Point(10, 4), Point(20, 20), Point(0, 20)], 2)
solutions = []
model = Solver(instance)
solution = model.solve()
assert solution.is_feasible(0.001)
```

```python
import matplotlib.pyplot as plt
from drone_range_assignment import plot
solutions = [solution]
fix, ax = plt.subplots(1,1)
plot(ax, solution)
ax.set_title(f"{len(instance.drones)} drone(s) \nPower: {round(solution.objective(),2)}")
plt.tight_layout()
plt.show()
```
